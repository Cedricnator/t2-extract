```python
import pandas as pd

df = pd.read_csv('../../data/top_repositories.csv')

df_cleaned = df.dropna()
df_cleaned
```

```python
print(f"Total de repositorios: {len(df_cleaned)}")

df_cleaned.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

language_counts = df_cleaned['language'].value_counts().head(5)
plt.figure(figsize=(12, 9))
sns.barplot(x=language_counts.index, y=language_counts.values)
plt.title('Top 5 de Lenguajes de Programacion en Repositorios Populares')
plt.xlabel('Numero de repositorios')
plt.ylabel('Lenguaje de Programacion')
plt.tight_layout()
plt.show()
```

```python
# Convertir columnas de fechas a datetime
df_cleaned['created_at'] = pd.to_datetime(df_cleaned['created_at'])
df_cleaned['updated_at'] = pd.to_datetime(df_cleaned['updated_at'])

# Extraer a√±o de creaci√≥n
df_cleaned['created_year'] = df_cleaned['created_at'].dt.year

# Contar por a√±o
year_counts = df_cleaned['created_year'].value_counts().sort_index()

# Visualizaci√≥n
plt.figure(figsize=(14, 8))
sns.lineplot(x=year_counts.index, y=year_counts.values)
plt.title('N√∫mero de Repositorios Populares Creados por A√±o')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de Repositorios')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Variables num√©ricas para correlaci√≥n
numeric_cols = ['stargazers_count', 'watchers', 'forks_count', 'open_issues_count']

# Matriz de correlaci√≥n
corr_matrix = df_cleaned[numeric_cols].corr()

# Visualizaci√≥n
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlaci√≥n entre M√©tricas de Repositorios')
plt.tight_layout()
plt.show()
```

```python
# Top 20 repositorios m√°s populares
top_repos = df_cleaned.sort_values('stargazers_count', ascending=False).head(20)

plt.figure(figsize=(14, 10))
sns.barplot(x='stargazers_count', y='name', data=top_repos)
plt.title('Top 20 Repositorios por N√∫mero de Estrellas')
plt.xlabel('N√∫mero de Estrellas')
plt.ylabel('Nombre del Repositorio')
plt.tight_layout()
plt.show()
```

```python
# Relaci√≥n entre estrellas e issues
plt.figure(figsize=(12, 8))
sns.scatterplot(x='stargazers_count', y='open_issues_count', 
                hue='language', size='forks_count',
                sizes=(20, 200), alpha=0.7, data=df_cleaned)
plt.title('Estrellas vs. Issues Abiertas')
plt.xlabel('N√∫mero de Estrellas')
plt.ylabel('N√∫mero de Issues Abiertas')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Agregar columna para ratio fork/estrella
df_cleaned['fork_star_ratio'] = df_cleaned['forks_count'] / df_cleaned['stargazers_count']

# Por lenguaje
lang_ratios = df_cleaned.groupby('language')['fork_star_ratio'].mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x=lang_ratios.values, y=lang_ratios.index)
plt.title('Proporci√≥n Fork/Estrella por Lenguaje')
plt.xlabel('Ratio Fork/Estrella (promedio)')
plt.ylabel('Lenguaje')
plt.tight_layout()
plt.show()
```

```python
# Add this cell to your notebook
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select features for clustering
features = ['stargazers_count', 'forks_count', 'open_issues_count', 'watchers']

# Get data and handle missing values
X = df_cleaned[features].fillna(0)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate WCSS (Within-Cluster Sum of Square) for different values of k
wcss = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(12, 6))
plt.plot(k_range, wcss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.grid(True, alpha=0.3)
plt.xticks(k_range)


plt.tight_layout()
plt.show()
```

```python
# Crear un diccionario con todos los an√°lisis para exportar a JSON
import json
import numpy as np

# Funci√≥n auxiliar para manejar tipos no serializables por defecto en JSON
def json_serializable(obj):
   if isinstance(obj, (np.integer, np.int64)):
      return int(obj)
   elif isinstance(obj, np.floating):
      return float(obj)
   elif isinstance(obj, np.ndarray):
      return obj.tolist()
   elif isinstance(obj, pd.Timestamp):
      return obj.strftime('%Y-%m-%d')
   else:
      return str(obj)

# Estructura de datos para el frontend
frontend_data = {
   "summary": {
      "total_repositories": len(df_cleaned),
      "avg_stars": df_cleaned['stargazers_count'].mean(),
      "avg_forks": df_cleaned['forks_count'].mean(),
      "avg_issues": df_cleaned['open_issues_count'].mean(),
   },
    
   "languages": {
      "labels": language_counts.index.tolist(),
      "values": language_counts.values.tolist()
   },
   
   "yearly_creation": {
      "years": year_counts.index.tolist(),
      "counts": year_counts.values.tolist()
   },
   
   "correlations": {
      "metrics": numeric_cols,
      "matrix": corr_matrix.to_dict()
   },
   
   "top_repositories": top_repos[['name', 'full_name', 'html_url', 'stargazers_count', 'forks_count', 'language']].to_dict('records'),
   
   "fork_star_ratio": {
      "languages": lang_ratios.index.tolist(),
      "ratios": lang_ratios.values.tolist()
   }
}
```

```python
from kneed import KneeLocator

# Find the elbow point automatically
knee = KneeLocator(list(k_range), wcss, curve="convex", direction="decreasing")
optimal_k_from_elbow = knee.elbow if knee.elbow else 3  # Fallback to 3 if no clear elbow

print(f"Automatically detected elbow point: k={optimal_k_from_elbow}")

# Apply K-means with optimal number of clusters (replace n with your chosen value)
optimal_k = 3  # Replace with the value you identify from the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_cleaned['cluster'] = kmeans.fit_predict(X_scaled)

# Visualize clusters
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
   df_cleaned['stargazers_count'], 
   df_cleaned['forks_count'],
   c=df_cleaned['cluster'], 
   cmap='viridis', 
   alpha=0.6, 
   s=100
)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Stars (log scale)')
plt.ylabel('Number of Forks (log scale)')
plt.title('Repository Clusters')
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Analyze cluster characteristics
cluster_analysis = df_cleaned.groupby('cluster')[features].mean()
print("Cluster Centers:")
print(cluster_analysis)

# Add cluster info to your JSON export
frontend_data["clusters"] = {
   "count": optimal_k,
   "repositories_per_cluster": df_cleaned['cluster'].value_counts().to_dict(),
   "cluster_centers": cluster_analysis.to_dict()
}

# Format elbow method data for the frontend
frontend_data["elbow_method"] = {
    "optimal_k": int(optimal_k_from_elbow),
    "points": [{"k": int(k), "wcss": float(w)} for k, w in zip(k_range, wcss)]
}

# Guardar datos para el frontend
with open('../../data/github_analysis.json', 'w', encoding='utf-8') as f:
   json.dump(frontend_data, f, default=json_serializable, indent=2)

print("Datos exportados correctamente para el frontend")
```
